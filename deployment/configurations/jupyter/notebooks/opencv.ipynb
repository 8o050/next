# Testing OpenCV

Just a simple `OpenCV` test to prove it is configured properly in Jupyter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('messi5.jpg')

convert = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(convert, cv2.COLOR_BGR2RGB)

plt.imshow(gray)
plt.title('A picture')
plt.show()
```
